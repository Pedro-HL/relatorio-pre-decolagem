
# 🚀 Relatório Operacional de Pré-Decolagem

Simulação de sistema automatizado de decisão baseado em telemetria aeroespacial.



## 📡 Telemetria Monitorada

- Temperatura interna e externa  
- Integridade estrutural  
- Nível de energia  
- Pressão dos tanques  
- Status dos módulos críticos  


```

class SistemaPreDecolagem:
    
    def __init__(self, temp_int, temp_ext, integridade, energia, pressao, modulos):
        self.temp_int = temp_int
        self.temp_ext = temp_ext
        self.integridade = integridade
        self.energia = energia
        self.pressao = pressao
        self.modulos = modulos

    def verificar_temperatura(self):
        return 18 <= self.temp_int <= 30 and -50 <= self.temp_ext <= 50

    def verificar_integridade(self):
        return self.integridade == 1

    def verificar_energia(self):
        return self.energia >= 70

    def verificar_pressao(self):
        return 4 <= self.pressao <= 6

    def verificar_modulos(self):
        return self.modulos == "ativos"

    def status_final(self):
        verificacoes = {
            "Temperatura": self.verificar_temperatura(),
            "Integridade": self.verificar_integridade(),
            "Energia": self.verificar_energia(),
            "Pressão": self.verificar_pressao(),
            "Módulos": self.verificar_modulos()
        }

        for item, status in verificacoes.items():
            print(f"{item}: {'OK' if status else 'FALHA'}")

        return "🚀 PRONTO PARA DECOLAR" if all(verificacoes.values()) else "❌ DECOLAGEM ABORTADA"


# Simulação de dados
sistema = SistemaPreDecolagem(
    temp_int=24,
    temp_ext=-12,
    integridade=1,
    energia=82,
    pressao=5.4,
    modulos="ativos"
)

print("\nStatus Final:", sistema.status_final())
```

```

# Análise Energética

capacidade_total = 1000  # kWh
carga_percentual = 0.82
consumo_decolagem = 300  # kWh
perdas_percentual = 0.05

energia_disponivel = capacidade_total * carga_percentual
perdas = energia_disponivel * perdas_percentual
energia_real = energia_disponivel - perdas
energia_restante = energia_real - consumo_decolagem

print("Energia disponível:", energia_disponivel, "kWh")
print("Perdas estimadas:", perdas, "kWh")
print("Energia real:", energia_real, "kWh")
print("Energia restante após decolagem:", energia_restante, "kWh")

if energia_restante > 0:
    print("Autonomia suficiente para operação segura.")
else:
    print("Energia insuficiente para decolagem segura.")
```
